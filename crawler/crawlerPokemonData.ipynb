```python
# -*- encoding: utf8-*-
import requests
from bs4 import BeautifulSoup
import csv
import time
import json
```

```python
user_agent = 'Mozillahtm/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'
headers = {'User-Agent': user_agent,
        'server': 'Wikipidia/1.0.4',
        'Referer': 'https://wiki.52poke.com/'}
wikiurl = "https://wiki.52poke.com/zh-hant/%E5%AE%9D%E5%8F%AF%E6%A2%A6%E5%88%97%E8%A1%A8%EF%BC%88%E6%8C%89%E5%85%A8%E5%9B%BD%E5%9B%BE%E9%89%B4%E7%BC%96%E5%8F%B7%EF%BC%89"
ablturl = "https://pokemondb.net/pokedex/all"
CPListURL = "https://gamepress.gg/sites/default/files/aggregatedjson/pokemon-data-full-en-PoGO.json?11412789301181126130"
```

```python
chineseName = []
attribute = []
EnglishName = []
PokemonABL = []
ABLNAME = []
outputList = []
weights = []
heights = []
MaxCpList = []
```

```python
def collectList():
    res = requests.get(url=wikiurl, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser');
    con = soup.findAll("div", class_="mw-parser-output")[-1]
    tables = con.findAll("table")[:-1]
    cnt = 0
    for table in tables:
        trs = table.findAll("tr")[1:];
        for tr in trs:
            startime = time.time()
            infos = tr.findAll("a")
            strinfo = []
            if  len(infos)==0:
                continue;
            for info in infos:
                strinfo.append(info.text)
            if len(strinfo[0])==0: 
                strinfo = strinfo[1:]
                infos = infos[1:]
            chineseName.append(strinfo[0])
            attribute.append(strinfo[1:])

            tds = tr.findAll("td")
            for td in tds:
                cxt = td.text
                if len(cxt)==0:
                    continue;
                elif cxt[0]>='A' and cxt[0]<='Z':
                    EnglishName.append(cxt[:-1])
                    break;
            
            href = infos[0].get("href")
            pokeurl = "https://wiki.52poke.com"+href
            respokemon = requests.get(url=pokeurl, headers=headers)
            soap = BeautifulSoup(respokemon.text, 'html.parser')
            try:
                heightboard = float(soap.findAll("table", class_="roundy bgwhite fulltable")[6].find("td").text[:-2])
                weightboard = float(soap.findAll("table", class_="roundy bgwhite fulltable")[7].find("td").text[:-3])
            except:
                del chineseName[-1]
                del attribute[-1]
                del EnglishName[-1]
                continue
            heights.append(heightboard)
            weights.append(weightboard)
            cnt = cnt+1
            now = time.time()
            print("Collected %d %s\t%.1fm %.1fkg\ttime:%.2fsec"
                %(cnt, chineseName[-1], heights[-1], weights[-1], now-startime))
            if not(cnt%100):
                print("collecting...")
                time.sleep(5)
```

```python
collectList()
```

```python
def collectABL():
    res = requests.get(url=ablturl, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")
    Pokedex = soup.find("div", class_= "resp-scroll").find("table", id="pokedex").find("tbody").findAll("tr")
    print("Collecting Ability Value...")
    for Pokemon in Pokedex:
        ABL = Pokemon.findAll("td", class_="cell-num")[1:]
        A = []
        for value in ABL:
            A.append(int(value.text))
        PokemonABL.append(A);
        name = Pokemon.find("td", class_="cell-name").text
        if name != None:
            ABLNAME.append(name)
    print("Collect Completed~~~")
```

```python
collectABL()
```

```python
def match():
    for poke in range(len(EnglishName)):
        for abl in range(len(ABLNAME)):
            if EnglishName[poke] == ABLNAME[abl]:
                outputList.append([chineseName[poke], EnglishName[poke]])
                
                if len(attribute[poke]) == 1:
                    outputList[-1].extend([attribute[poke][0], attribute[poke][0]])
                else:
                    outputList[-1].extend([attribute[poke][0], attribute[poke][1]])
                outputList[-1].extend([heights[poke], weights[poke]])
                outputList[-1].extend(PokemonABL[abl])
                print("matched:", outputList[-1])
                break
    print("total:%d"%len(outputList))
    print("Matching Complete(^_^!)")
```

```python
match()
```

```python
def save_as_csv():
    print(len(outputList))
    print("Saving Data...")
    with open("Pokemon.csv", 'w', newline='', encoding="utf-8") as fp:
        writer = csv.writer(fp, delimiter='\t')
        writer.writerow(['Name', 'English','Type1', 'Type2', 'Height(m)', 'Weight(kg)', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'])
        for line in outputList:
            print (line)
            writer.writerow(line)
    print("Save Complete!!!")
```

```python
save_as_csv()
```

```python
def getMaxCP():
    print("Fetching Max CP list...")
    res = requests.get(url=CPListURL, headers=headers)
    
    board = json.loads(res.text)
    
    for pokemon in board:
        
        name = pokemon['title_1']
        name = BeautifulSoup(name)
        name = name.text
        
        maxcp = float(pokemon['lvl40'])
        print(name, maxcp)
        for out in outputList:
            if out[1] == name:
                MaxCpList.append(out)
                MaxCpList[-1].append(maxcp)
                outputList.remove(out)
                
    
    print("Fetch Completed")
```

```python
getMaxCP()
```

```python
def save_as_csv_with_cp():
    print("Saving Data...")
    with open("Pokemon-maxcp.csv", 'w', newline='', encoding="utf-8") as fp:
        writer = csv.writer(fp, delimiter=',')
        writer.writerow(['Name', 'English','Type1', 'Type2', 'Height(m)', 'Weight(kg)', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'max-cp'])
        for line in MaxCpList:
            writer.writerow(line)
    with open("Pokemon-without.csv", 'w', newline='', encoding="utf-8") as fp:
        writer = csv.writer(fp, delimiter=',')
        writer.writerow(['Name', 'English','Type1', 'Type2', 'Height(m)', 'Weight(kg)', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'])
        for line in outputList:
            writer.writerow(line)
    print("Save Complete!!!")
```

```python
save_as_csv_with_cp()
```
